```python
import pandas as pd

data = pd.read_csv('/content/pengeluaran_perkapita.csv')

data.head()
```

```python
data = data.drop_duplicates()
```

```python
import numpy as np


```

```python
list(data.columns)
```

```python
target_numerik = 'gk.total.perdesaanperkotaan.2022.september'

data[target_numerik] = pd.to_numeric(
    data[target_numerik], errors='coerce'
)

data[[ 'provinsi', target_numerik ]].head()
```

```python
data['kelas_pengeluaran'] = pd.qcut(
    data[target_numerik],
    q=3,
    labels=['Rendah', 'Sedang', 'Tinggi']
)

data[['provinsi', target_numerik, 'kelas_pengeluaran']].head()
```

```python
data['kelas_pengeluaran'].value_counts()
```

```python
fitur = [
    'gk.total.perdesaanperkotaan.2020.september',
    'gk.total.perdesaanperkotaan.2021.september',
    'gk.total.perdesaanperkotaan.2022.maret',
    'gk.total.perdesaanperkotaan.2022.september'
]

X = data[fitur]
y = data['kelas_pengeluaran']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(
    n_neighbors=5,
    metric='euclidean'
)

knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(
    cm, annot=True, fmt='d',
    cmap='Blues',
    xticklabels=knn.classes_,
    yticklabels=knn.classes_
)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix KNN')
plt.show()
```
